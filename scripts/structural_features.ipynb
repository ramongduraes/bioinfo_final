# Parsing PDB files to get structural features

```python
import os
from Bio import PDB
```

```python
os.listdir('../data')
```

```python
test_pdb = '../data/P0DTC1.pdb'
```

```python
parser = PDB.PDBParser()
```

```python
# pdb1 ='./4twu.pdb' 
structure = parser.get_structure("P0DTC1", test_pdb) 
```

```python
model = structure[0] 
chain = model['A'] 
```

```python
chain.__dict__
```

```python
residue1 = chain[1] 
residue2 = chain[2]
atom1 = residue1['CA'] 
atom2 = residue2['CA'] 

distance = atom1-atom2 

print(distance)
```
