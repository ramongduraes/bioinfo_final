# Features de aminoácidos

```python
import pandas as pd
import pickle
```

### Aminoácidos

```python
df = pd.read_pickle("../data/aa_props_dataframe.pkl")
df["polar"] = df["class"].apply(lambda x: 0 if "Non-polar" in x or "Aromatic" in x else 1)
amino_dict = df.to_dict()
df.head(20)
```

### Mutações

```python
with open('../data/swissmodel_dataframe.pkl', 'rb') as f:
    mutations_df = pickle.load(f)
    
mutations_df['LeftPol'] = mutations_df["From"].apply(lambda x: amino_dict["polar"][x])
mutations_df['RightPol'] = mutations_df["To"].apply(lambda x: amino_dict["polar"][x])
mutations_df['LeftVol'] = mutations_df["From"].apply(lambda x: amino_dict["volume"][x])
mutations_df['RightVol'] = mutations_df["To"].apply(lambda x: amino_dict["volume"][x])
mutations_df['LeftCharge'] = mutations_df["From"].apply(lambda x: amino_dict["charge_at_pH"][x])
mutations_df['RightCharge'] = mutations_df["To"].apply(lambda x: amino_dict["charge_at_pH"][x])
mutations_df['LeftHidro'] = mutations_df["From"].apply(lambda x: amino_dict["gravy"][x])
mutations_df['RightHidro'] = mutations_df["To"].apply(lambda x: amino_dict["gravy"][x])
mutations_df['LeftIsoEl'] = mutations_df["From"].apply(lambda x: amino_dict["isoelectric_point"][x])
mutations_df['RightIsoEl'] = mutations_df["To"].apply(lambda x: amino_dict["isoelectric_point"][x])

mutations_df = mutations_df.iloc[:,6:]
mutations_df.to_csv("../data/amino_mut.csv", index=False)
mutations_df.head()
```
